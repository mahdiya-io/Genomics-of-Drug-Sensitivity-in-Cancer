```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
gdsc_data_cleaned = pd.read_csv("gdsc_data_preprocessed.csv")
X_test = pd.read_csv("X_test.csv")
X_train = pd.read_csv("X_train.csv")
Y_test = pd.read_csv("y_test.csv")
Y_train = pd.read_csv("y_train.csv")
```

```python
# converts dp to np array and then flattens it
y_test = Y_test.to_numpy().ravel()
y_train = Y_train.to_numpy().ravel()
```

# Testing the models



## Logistic Regression



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)
logistic_predictions = logistic_model.predict(X_test)
logistic_score = accuracy_score(y_test, logistic_predictions)
print("Logistic Regression Model Accuracy: ", logistic_score)
print("Logistic Regression Model Precision: ", sklearn.metrics.precision_score(y_test, logistic_predictions, average='micro'))
print("Logistic Regression Model Recall: ", sklearn.metrics.recall_score(y_test, logistic_predictions, average='micro'))
print("Logistic Regression Model F1 Score: ", sklearn.metrics.f1_score(y_test, logistic_predictions, average='micro'))
```

```python
logistic_confusion_matrix = sklearn.metrics.confusion_matrix(y_test, logistic_predictions)
sklearn.metrics.ConfusionMatrixDisplay(logistic_confusion_matrix).plot()
plt.show
```

## Ridge Classifier

```python
from sklearn.linear_model import RidgeClassifier

ridge_model = RidgeClassifier()
ridge_model.fit(X_train, y_train)
ridge_predictions = ridge_model.predict(X_test)
ridge_score = accuracy_score(y_test, ridge_predictions)
print("Ridge Classifier Model Accuracy: ", ridge_score)
print("Ridge Classifier Model Precision: ", sklearn.metrics.precision_score(y_test, ridge_predictions, average='micro'))
print("Ridge Classifier Model Recall: ", sklearn.metrics.recall_score(y_test, ridge_predictions, average='micro'))
print("Ridge Classifier Model F1 Score: ", sklearn.metrics.f1_score(y_test, ridge_predictions, average='micro'))
```

```python
# Ridge Classifier Confusion Matrix
ridge_confusion_matrix = sklearn.metrics.confusion_matrix(y_test, ridge_predictions)
sklearn.metrics.ConfusionMatrixDisplay(ridge_confusion_matrix).plot()
plt.show
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

forest_model = RandomForestClassifier()
forest_model.fit(X_train, y_train)
forest_predictions = forest_model.predict(X_test)
forest_score = accuracy_score(y_test, forest_predictions)
print("Random Forest Classifier Model Accuracy: ", forest_score)
print("Random Forest Classifier Model Precision: ", sklearn.metrics.precision_score(y_test, forest_predictions, average='micro'))
print("Random Forest Classifier Model Recall: ", sklearn.metrics.recall_score(y_test, forest_predictions, average='micro'))
print("Random Forest Classifier Model F1 Score: ", sklearn.metrics.f1_score(y_test, forest_predictions, average='micro'))
```

```python
# Random Forest Classifier Confusion Matrix
forest_confusion_matrix = sklearn.metrics.confusion_matrix(y_test, forest_predictions)
sklearn.metrics.ConfusionMatrixDisplay(forest_confusion_matrix).plot()
plt.show
```

# K-fold Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
dropped_columns = ['LN_IC50', 'COSMIC_ID']
X = gdsc_data_cleaned.drop(columns=dropped_columns, axis=1)
y = gdsc_data_cleaned['LN_IC50']
```

```python
# random forest classifier
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
forest_k_fold_score = cross_val_score(forest_model, X, y, cv=kfold)
print("Random Forest Classifier Model Accuracy: ", forest_k_fold_score.mean())
```

```python
# KNN
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
knn_k_fold_score = cross_val_score(knn_model, X, y, cv=kfold)
print("KNN Model Accuracy: ", knn_k_fold_score.mean())
```

```python
# decision tree classifier
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
decision_k_fold_score = cross_val_score(decision_model, X, y, cv=kfold)
print("Decision Tree Model Accuracy: ", decision_k_fold_score.mean())
```
